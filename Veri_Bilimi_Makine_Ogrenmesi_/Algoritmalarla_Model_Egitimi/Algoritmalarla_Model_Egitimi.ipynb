```python
veri = pd.get_dummies(veri, columns=['Cinsiyet','Yaş Grubu','Medeni Durum','Eğitim Düzeyi','Yaşadığı Şehir', 'İstihdam Durumu','Eğitime Devam Etme Durumu', 'En Çok İlgilendiği Ürün Grubu'])
```

```python
veri
```

```python
y=veri["Öbek İsmi"].values
x_data=veri.drop(["Öbek İsmi"],axis=1)
```

```python
#normalizasyon
x=(x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
#test bölünmesi
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=1)
```

K - en yakın komşu algoritması

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
komsu_sayisi= 4
knn=KNeighborsClassifier(n_neighbors=komsu_sayisi)
knn.fit(x_train,y_train)

prediction = knn.predict(x_test)
print("{} En Yakın Komşu Test Doğruluk : {}".format(komsu_sayisi,knn.score(x_test,y_test)))
```

Destek Vektör Makinesi algoritması

```python
from sklearn.svm import SVC
svm = SVC(random_state=1)
svm.fit(x_train,y_train)
print("doğruluk: {}".format(svm.score(x_test,y_test)))
```

Karar Ağacı

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("doğruluk: {}".format(dt.score(x_test,y_test)))
```

Rastgele Orman= Bir çok karar ağacının toplamından oluşur. karar ağaçlarının her birinin verdiği sonuçlara göre çoğunluk değeri döndürür.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,random_state=1) #100 Ağaç
rf.fit(x_train,y_train)
print("doğruluk:{}".format(dt.score(x_test,y_test)))
```

confusion matrix: doğru ve yanlış tahminleri gösteren matrix.


```python
from sklearn.metrics import confusion_matrix
y_pred = rf.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
cm #(0,0),(1,1),(2,2)... doğru sonuçlar, gerisi yanlış.
```
