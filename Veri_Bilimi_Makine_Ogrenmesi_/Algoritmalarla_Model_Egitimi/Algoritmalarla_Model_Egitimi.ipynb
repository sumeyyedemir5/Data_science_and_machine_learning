```python
veri = pd.get_dummies(veri, columns=['Cinsiyet','YaÅ Grubu','Medeni Durum','EÄitim DÃ¼zeyi','YaÅadÄ±ÄÄ± Åehir', 'Ä°stihdam Durumu','EÄitime Devam Etme Durumu', 'En Ãok Ä°lgilendiÄi ÃrÃ¼n Grubu'])
```

```python
veri
```

```python
y=veri["Ãbek Ä°smi"].values
x_data=veri.drop(["Ãbek Ä°smi"],axis=1)
```

```python
#normalizasyon
x=(x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
#test bÃ¶lÃ¼nmesi
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=1)
```

K - en yakÄ±n komÅu algoritmasÄ±

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
komsu_sayisi= 4
knn=KNeighborsClassifier(n_neighbors=komsu_sayisi)
knn.fit(x_train,y_train)

prediction = knn.predict(x_test)
print("{} En YakÄ±n KomÅu Test DoÄruluk : {}".format(komsu_sayisi,knn.score(x_test,y_test)))
```

Destek VektÃ¶r Makinesi algoritmasÄ±

```python
from sklearn.svm import SVC
svm = SVC(random_state=1)
svm.fit(x_train,y_train)
print("doÄruluk: {}".format(svm.score(x_test,y_test)))
```

Karar AÄacÄ±

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("doÄruluk: {}".format(dt.score(x_test,y_test)))
```

```python
from sklearn.metrics import confusion_matrix
y_pred = rf.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
cm #(0,0),(1,1),(2,2)... doÄru sonuÃ§lar, gerisi yanlÄ±Å.
```
