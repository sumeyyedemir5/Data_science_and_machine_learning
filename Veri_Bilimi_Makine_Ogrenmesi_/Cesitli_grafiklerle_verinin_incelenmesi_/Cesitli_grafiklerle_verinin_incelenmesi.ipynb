```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri=pd.read_csv("olimpiyatlar.csv")
veri.head()
```

Saçılım Grafiği

```python
def sacilimGrafigiCiz():
   sns.scatterplot(x='boy',y='kilo',data=veri)
   plt.title('Boy ve Ağırlık İlişkisi')
   plt.show()

sacilimGrafigiCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafigiCiz()
```

```python
sns.scatterplot(x='boy',y='kilo',hue='madalya',data=veri)
plt.title('Madalyaya göre boy-kilo dağılımı')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='boy',y='kilo',hue='cinsiyet',style='madalya',size='yaş',sizes=(15,200),data=veri)
plt.xlabel('Sporcuların Boyu')
plt.ylabel('Sporcuların kiloları')
plt.title('Cinsiyete göre boy-kilo dağılımı')
plt.show()
```

Renk Paletleri

```python
sns.scatterplot(x='boy',y='kilo',hue="madalya",data=veri,palette="Set1")
plt.title('madalya')
plt.show()
```

```python
sns.scatterplot(x='boy',y='kilo',hue="madalya",data=veri,palette="rocket")
plt.title('madalya')
plt.show()
```

Çizgi Grafiği

```python
sns.lineplot(x='boy',y='kilo',hue='cinsiyet',data=veri)
plt.title('cinsiyete göre boy kilo ilişkisi')
plt.show()
```

Histogram

```python
sns.displot(veri,x='kilo',hue="cinsiyet")
plt.ylabel("Frekans")
plt.show()
```

```python
sns.catplot(x="madalya",y="yaş",hue="cinsiyet",col="sezon",data=veri,kind="bar",height=4,aspect=0.7)
plt.show()
```

**kutu grafiği**

```python
sns.boxplot(x="sezon",y="boy",hue="cinsiyet",data=veri,palette="Set2")
plt.show()
```

```python
veri_gecici=veri.loc[:,["yaş","kilo","boy"]]
sns.boxplot(data=veri_gecici,orient="h",palette="Set2")
plt.show()
```

```python
sns.heatmap(veri.corr(),annot=True,linewidths=0.5,fmt='.1f')
plt.show()
```

```python
sns.countplot(x="sehir",data=veri)
plt.xticks(rotation=90)
plt.show()
```
